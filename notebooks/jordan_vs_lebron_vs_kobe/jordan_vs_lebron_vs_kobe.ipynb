## Jordan vs LeBron vs Kobe
### **Introdução**
Muitos apesctos devem ser levados em conta para determinar o maior atleta de todos os tempos em qualquer modalidade. No contexto de esportes coletivos, a avalição se torna ainda mais complexa pelo simples fato dessas atividades demandarem mais de um indíviduo por time para haver uma disputa e os rendimentos individuais de qualquer atleta serem influenciados positiva e/ou negativamente de acordo com o desempenho de seus companheiros. Outras questões como a época em que o jogador jogou ou em qual local tem forte influência em suas estatísticas. 

Jogadores de épocas diferentes, por exemplo, muitas vezes são incoparáveis tanto por conta da evolução da condição física no esporte quanto por conta de mudanças nas regras e táticas/estilo de jogo. 

Um atleta que já é naturalmente superior em uma liga competitiva, ao se transferir para uma liga com um nível técnico mais baixo ou até jogar em um time com investimento e/ou um elenco muito superior aos demais provavelmente terá números superiores.

Esses e outros aspectos externos existem e caso não sejam levados em conta, podem tendeciar a decisão de um ranking.

Apesar disso tudo, o que é indiscutível, não só no mundo dos esportes, são dados. Números não mentem. A análise a seguir será exclusivamente baseada em estatísticas para determinar o maior jogador de basquete de todos os tempos. 

Felizmente, alguns dos aspectos externos citados anteriormente que são difíceis de quantificar tem uma influência pequena nesta análise. O temporal, por exemplo, apesar de existir, impacta pouco. Jordan disputou 4 temporadas em que Kobe já jogava. Kobe e Lebron diviram os holofotes da NBA por 13 anos. Além disso, os 3 jogaram por times competitivos da NBA, como mostram o número de títulos de cada um. Apesar de outros pontos influenciarem no rendimento de cada um (lesões, por exemplo), eles serão deixados de lado para simplificar a análise.

#### **Cálculo da nota final**

Para determinar "o maior de todos os tempos", uma nota será dada para cada um a fim de indicar objetivamente o melhor dos 3.

O cálculo será divido em três partes: títulos, que são uma medida bastante simples do impacto coletivo de cada jogador; prêmios individuais, medindo o impacto individual; e estatísticas individuais, que servem como indicador primário da capacidade de cada um de dominar o jogo em aspectos ofensivos e defensivos.

Para calcular o valor final, cada parte tem um peso. Os títulos representarão 20% da nota, pois, como dito anteriormente, tem uma dependência muito grande do aspecto coletivo do jogo; prêmios individuais, 15%; e as estatísticas em si, 75%, pois são a forma mais objetiva de avaliar a capacidade de cada um e não dependem de nehuma forma de avaliação subjetiva para serem atribuídas, diferentemente dos prêmios indiviuais. 

Para cada tópico avaliado em cada parte, um determinado número de pontos serão distribuídos, de acordo com a classificação de cada um no critério avalidado.

Para os títulos, serão 10 pontos para o primeiro colocado, 7 para o segundo e 4 para o terceiro.

Para os prêmios, serão 3 pontos paro o primeiro colocado, 2 para o segundo e 1 para o terceiro.

Para as estatísticas, para a análise de total, total por jogo e total por temporada, o primeiro colocado recebe 15 pontos, o segundo 10 e o terceiro 5. Para as duas métricas de presença no Top 5, 5 pontos serão dados para o primeiro, 4 para o segundo e assim sucessivamente.

Em caso de empate, a nota mais alta entre os empatados será dada.

Ao final, a soma dos pontos de cada estatística serão somados e dividos por 5. Esse resultado somado com os pontos recebidos por prêmios e títulos dará uma nota entre 28 e 100 para cada jogador e por meio dela será decidido quem foi o maior de todos os tempos.

### Estatísticas individuais

A avaliação das estatíticas individuais tem o foco em cinco aspectos fundamentais para um jogador de basquete completo: arremesso (pontuação e precisão), assistências, rebotes ofensivos e defensivos e turnovers (erros que custam a posse de bola). Por meio desses 5 pontos, pode-se ter noção e avaliar o desempenho de cada atleta tanto no ataque, pontuando, servindo os companheiros de equipe e recuperando rebotes, tudo isso sem cometer error que custem a posse de bola para o time, quanto na defesa, com rebotes defensivos.

Cada um dos 5 pontos terão estatísticas específicas a serem analisadas. São elas:
* Arremesso: Total de pontos;
* Assistências: Total de assistências;
* Rebotes ofensivos: Total de rebotes ofensivos;
* Rebotes defensivos: Total de rebotes defensivos;
* Turnoves: Total de turnoves.

Para cada uma das 5 estatísticas, 5 medidas serão analisadas, visando avaliar não só os picos de performance, mas a consistência ao longo dos anos também:
* Total
* Média por jogo;
* Média por temporada;
* Presença no Top 5 geral no total por temporada;
* Presença no Top 5 geral na média por jogo por temporada.

No caso dos turnovers, a regra seguirá a mesma lógica, mas com a pontuação dada de maneira invertida.

A fonte de todos os dados é do site [Basketball Reference](https://www.basketball-reference.com/)

Usando exclusivamente os critérios explicados acima em cada seção, Michael Jordan pode ser considerado o maior dentre os 3 jogadores. Apesar dessa ser uma análise simples e não envolver outros jogadores fora os três, ela mostra que Michael, além de ter números superiores em diversos critérios, manteve uma consistência grande ao longo dos anos. Outro ponto que engrandece ainda mais seus feitos é o fato dele ser o jogador entre os três com menos jogos (quase 300 a menos que Kobe e 200 a menos que Lebron) e menos temporadas disputadas.

Para efeito de comparação, dos totais e das médias (por jogo e temporada) avaliadas para as 5 estastísticas, totalizando 15 medidas, ele fica em primeiro em 8. Em outras 4, em segundo. Nas três estatíticas que Jordan fica em terceiro (total de pontos feitos, total de assistências e total de rebotes defensivos), ele é 5º, 45º e 79º, respectivamente, de toda história na NBA.

Fora isso, nas 3 temporadas em que ele está bem abaixo dos números apresentados por Kobe e James, bastante jogos não contaram com sua presença. Na segunda temporada de sua carreira (1985-1986), ele perdeu 64 jogos, por conta de um pé quebrado no terceiro jogo da temporada; no que seria sua décima temporada (1993-1994) não jogou pois antes de seu início, anunciou sua primeira aposentadoria, bastante influenciada pela morte de seu pai, e passou o ano jogando beisebol. Ao retornar para NBA na temporara 1994-1995, Jordan jogou apenas 17 jogos. Porém marcou 457 pontos, uma média de quase 27 pontos por jogo; já em seu penúltimo ano na liga, jogou apenas 60 jogos por conta de um problema no joelho direito e marcou uma média de 23 pontos por jogo, a terceira menor de sua carreira.
